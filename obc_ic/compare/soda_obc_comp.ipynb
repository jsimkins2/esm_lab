```python
import xarray as xr

soda3temp = xr.open_dataset("temp_001_1993_soda3.3.1.nc")
soda12temp = xr.open_dataset("temp_001_1993.nc")
```

```python
soda3temp
```

```python
soda3temp.temp_segment_001.isel(time=0, ny_segment_001=0).plot()
```

```python
soda12temp.temp_segment_001.isel(time=0, ny_segment_001=0).plot()
```

```python
delta = soda3temp.temp_segment_001.isel(time=0, ny_segment_001=0).values - soda12temp.temp_segment_001.isel(time=0, ny_segment_001=0).values
```

```python
delta
```

```python
import numpy as np
np.mean(delta)
```
