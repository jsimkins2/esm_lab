# Change _FillValue and/or Missing Value for netCDF file

This is necessary to do for initial conditions files for MOM6. _FillValue must be set to `-1.e+20`, same with missing value. Note that this code only works with newer versions of xarray. Also, I've only had success running this on Unix machines (it worked on poseidon, but not on triton)

```python

import xarray as xr

soda = xr.open_dataset("/Users/james/Documents/Github/esm_lab/grid_gen/gridInfo/nwa25/ic/soda3.3.1_5dy_ocean_reg_2010_01_05.nc")
```

```python
for var in soda.data_vars:
    if soda[var].dims != ('time',):
        soda[var].attrs['missing_value'] = -1.e+20
        soda[var] = soda[var].fillna(-1.e+20)
        
```

```python
soda.to_netcdf("/Users/james/Documents/Github/esm_lab/grid_gen/gridInfo/nwa25/ic/soda3.3.1_5dy_ocean_reg_2010_01_05_xarray.nc", mode='w', format='NETCDF3_64BIT')
```
