```python

%matplotlib inline
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import numpy as np
import xarray as xr
import xesmf as xe
```

```python

ds = xr.tutorial.open_dataset('rasm') # use xr.tutorial.load_dataset() for xarray<v0.11.0
ds
```

```python
dr = ds['Tair']
dr
```

```python
plt.figure(figsize=(12,2));
ax = plt.axes(projection=ccrs.PlateCarree());
dr[0].plot.pcolormesh(ax=ax, x='xc', y='yc');
ax.coastlines();
```

```python
plt.scatter(ds['xc'], ds['yc'], s=0.01)  # plot grid locations
plt.ylim([-90, 90])
plt.xlabel('lon')
plt.ylabel('lat')
```

```python
ds = ds.rename({'xc': 'lon', 'yc': 'lat'})
ds
```

```python
ds_out = xe.util.grid_global(5, 4)
ds_out  # contains lat/lon values of cell centers and boundaries.
```

```python
regridder = xe.Regridder(ds, ds_out, 'bilinear')
dr_out = regridder(dr)
```

```python
dr_out
```

```python
plt.figure(figsize=(12,4));
ax = plt.axes(projection=ccrs.PlateCarree())
dr_out[0].plot.pcolormesh(ax=ax, x='lon', y='lat');
ax.coastlines();
```
