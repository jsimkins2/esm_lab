# Simple Grids

- x_c, which represents the cell center
- x_g, which represents the left cell edge


```python
import xarray as xr
import numpy as np

ds = xr.Dataset(coords={'x_c': (['x_c',], np.arange(1,10)),
                        'x_g': (['x_g',], np.arange(0.5,9))})


```

```python
ds
```

At this point, xarray has no idea that x_c and x_g are related to each other; they are subject to standard xarray broadcasting rules. When we create an xgcm.Grid, we need to specify that they are part of the same axis. We do this using the coords keyword argument, as follows:

```python
from xgcm import Grid
```

```python
grid = Grid(ds, coords={'X': {'center': 'x_c', 'left': 'x_g'}})
```

```python
grid
```

The arrows after each coordinate indicate the default shift positions for interpolation and difference operations: operating on the center coordinate (x_c) shifts to the left coordinate (x_g), and vice versa.

```python
ds = xr.Dataset(coords={'x_c': (['x_c',], np.arange(1,10), {'axis': 'X'}),
                       'x_g': (['x_g',], np.arange(0.5,9), {'axis': 'X', 'c_grid_axis_shift': -0.5})})
```

```python
ds
```

```python
grid = Grid(ds)
```

```python
grid
```

```python
f = np.sin(ds.x_c * 2*np.pi/9)
```

```python
f
```

```python
f.plot()
```

```python
f_interp = grid.interp(f, axis='X')
```

```python
f_interp
```

Now the data is output onto the g grid instead of the c grid!

```python
f_diff = grid.diff(f, axis='X')
```

```python
f_diff
```

```python
grid.cumsum(f_diff, 'X')
```
